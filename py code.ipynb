```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
bike= pd.read_csv("Downloads/day.csv")
```

```python
bike.head()
```

```python
bike.info()
```

```python
bike.describe()
```

```python
bike.columns
```

```python
#Dropping the columns 'casual', 'registered' because 'cnt'='casual'+ 'registered', we can drop 'dteday' column as we have 'yr' and 'mnth' columns
# instant is an index column,  so we can drop it
bike.drop(['casual','registered','dteday','instant'],axis=1,inplace=True)
bike.head()
```

```python
#by observing the datadictionary we can conclude that season,yr,mnth,holiday,weekday,workingday,weathersit are categorical variables
#We will replace season,weekday and weathersit with values
```

```python
bike['season'].replace({1:"spring",2:"summer",3:"fall",4:"winter"},inplace=True)
bike.head(10)
```

```python
bike['weathersit'].replace({1: "Clear, Few clouds, Partly cloudy, Partly cloudy", 2: "Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist", 3: "Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds", 4: "Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog"},inplace=True)
bike.head(10)
```

```python
bike['weekday'].replace({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"},inplace=True)
bike.head(10)
```

```python
#Pair plots of numerical variables
sns.pairplot(bike, vars=['temp','atemp','hum','windspeed',"cnt"])
plt.show()
```

```python
bike.corr()
```

```python
plt.figure(figsize = (16, 12))
sns.heatmap(bike.corr(), annot=True, cmap="Oranges")
plt.show()
```

```python
#droping the highly correlated values
#correlation between 'temp' and 'atemp' is 0.99 which is near to 1
#so drop temp and proceed with atemp

bike.drop(['temp'],axis=1,inplace=True)
bike.head()
```

```python
#Data visualization for Categorical columns
plt.figure(figsize=(20, 15))
plt.subplot(2,4,1)
sns.boxplot(x = 'season', y = 'cnt', data = bike)
plt.subplot(2,4,2)
sns.boxplot(x = 'yr', y = 'cnt', data = bike)
plt.subplot(2,4,3)
sns.boxplot(x = 'mnth', y = 'cnt', data = bike)
plt.subplot(2,4,4)
sns.boxplot(x = 'workingday', y = 'cnt', data = bike)
plt.subplot(2,4,5)
sns.boxplot(x = 'holiday', y = 'cnt', data = bike)
plt.subplot(2,4,6)
sns.boxplot(x = 'weekday', y = 'cnt', data = bike)
plt.subplot(2,4,7)
sns.boxplot(x = 'weathersit', y = 'cnt', data = bike)
plt.show()
```

```python
#CREATING DUMMY VARIABLES FOR CATEGORICAL DATA 

Season_dummyvar=pd.get_dummies(bike['season'],drop_first=True)
Weather_dummyvar=pd.get_dummies(bike['weathersit'],drop_first=True)
Day_of_week=pd.get_dummies(bike['weekday'],drop_first=True)
Month=pd.get_dummies(bike['mnth'],drop_first=True)
```

```python
bike=pd.concat([bike,Season_dummyvar],axis=1)
bike=pd.concat([bike,Weather_dummyvar],axis=1)
bike=pd.concat([bike,Day_of_week],axis=1)
bike=pd.concat([bike,Month],axis=1)
bike.info()
```

```python
#Deleting the orginal columns season.weathersit,weekday,mnth
bike.drop(['season'],axis=1,inplace=True)
bike.drop(['weathersit'],axis=1,inplace=True)
bike.drop(['weekday'],axis=1,inplace=True)
bike.drop(['mnth'],axis=1,inplace=True)
bike.head()
```

```python
#splitting the data to train and test data
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
bike_train, bike_test = train_test_split(bike, train_size = 0.7, test_size = 0.3, random_state = 100)
```
